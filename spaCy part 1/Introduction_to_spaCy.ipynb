The `with` statement is primarily used to simplify resource management and avoid resource leaks. It ensures that a resource is properly cleaned up after its usage, even if errors occur.

It works with objects that support the context management protocol (i.e., they have `__enter__` and `__exit__` methods). When the `with` block is entered, `__enter__` is called, and when the block is exited (either normally or due to an exception), `__exit__` is called.

### Example: File Handling with `with`

Without `with`, you'd typically write code like this to ensure a file is closed:

````python
file = open('example.txt', 'w')
try:
    file.write('Hello, world!')
finally:
    file.close()
````

Using `with`, the same operation is much cleaner and safer:

````python
with open('example.txt', 'w') as file:
    file.write('Hello, world from with!')

# The file is automatically closed when exiting the 'with' block

# You can verify the content
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
````

A token is an object itself, it also contains individual attributes like `.text`

- `token.text`: This captures the raw characters of the word, punctuation, or whitespace exactly as they were written before any processing (like lowercasing or lemmatization).

- `token.left_edge`: This attribute refers to the first token of the smallest syntactic unit (a phrase or clause) that contains the current token. It's useful for understanding the boundaries of grammatical units.

If part of a sequence of tokens that are collectively meaningful, known as multi-word tokens, this will tell us where the multi-word token begins.

- `token.right_edge`: Similar to `left_edge`, this attribute refers to the last token of the smallest syntactic unit that contains the current token. This will tell us where the multi-word token ends.

- `token.ent_type`: This provides the hash value of the named entity type. It's an integer ID representing the entity category.

- `token.ent_type_`: This is the string representation of the named entity type. For example, 'GPE' stands for Geopolitical Entity.

- `token.ent_iob_`: This indicates whether the token is inside, outside, or at the beginning of a named entity. 'I' for inside, 'O' for outside, 'B' for beginning. This is part of the IOB (Inside, Outside, Beginning) tagging scheme used in Named Entity Recognition.

- `token.lemma_`: This provides the base form of the word, also known as its lemma. For example, the lemma of 'running' is 'run', and the lemma of 'states' is 'states' (as it's already in its base form for a proper noun).
Base form of the token, with no inflectional suffixes

- `token.morph`: This attribute returns a `Morphology` object containing detailed morphological features of the token, such as number (singular/plural), tense, aspect, etc. For 'States', it indicates `Number=Sing` in this context, suggesting its role in the phrase 'United States'.

- `token.pos_`: This gives the coarse-grained part-of-speech tag. 'PROPN' stands for proper noun, indicating that 'States' in 'United States' is treated as a proper noun.

- `token.dep_`: This represents the syntactic dependency relation between the token and its head. 'nsubj' stands for nominal subject, indicating that 'States' is the subject of a clause or sentence.

- `token.lang_`: This attribute identifies the language of the token. 'en' indicates English.

# Named Entity Recognition

In spaCy, `doc.ents` signifies a tuple of named entities found in the `Doc` object. These entities are 'real-world objects' that a trained spaCy pipeline has identified in the text, such as persons, organizations, locations, dates, and more. Each entity is itself a `Span` object with various attributes like `text`, `label_` (the entity type), and `start_char`/`end_char` (the character offsets).

A 'trained spaCy pipeline' refers to a series of processing components that spaCy uses to analyze text. When you load a model like `en_core_web_sm` with `nlp = spacy.load("en_core_web_sm")`, you're loading a pre-trained pipeline.

Here's a breakdown:

1.  **Pipeline Components**: A spaCy pipeline consists of multiple steps (components) that perform different NLP tasks. Common components include:
    *   **Tokenizer**: Breaks text into words, punctuation, etc.
    *   **Tagger**: Assigns Part-of-Speech (POS) tags (e.g., noun, verb, adjective).
    *   **Parser**: Analyzes the grammatical structure of sentences (dependency parsing).
    *   **Named Entity Recognizer (NER)**: Identifies 'named entities' like people, organizations, locations, dates, etc.

2.  **Training**: These components, especially the NER, tagger, and parser, are powered by statistical models. 'Training' involves feeding these models vast amounts of annotated text data. For example, during NER training, the model learns to recognize patterns in text that correspond to different entity types by seeing many examples of correctly labeled entities.

3.  **Pre-trained Models**: When we download `en_core_web_sm`, we're getting a model that has already been trained on large English datasets. This means it has learned the statistical patterns to perform tasks like identifying entities without us having to explicitly train it ourselves. It's ready to use 'out-of-the-box' for general-purpose English text.

# Word Vectors
Word vectors, or word embeddings, are numerical representations of words in multidimensional space through matrices. The purpose of the word vector is to get a computer system to understand a word. Computers cannot understand text efficiently. They can, however, process numbers quickly and well. For this reason, it is important to convert a word into a number.

Initial methods for creating word vectors in a pipeline take all words in a corpus and convert them into a single, unique number. These words are then stored in a dictionary that would look like this: {“the”: 1, “a”, 2} etc. This is known as a bag of words. This approach to representing words numerically, however, only allow a computer to understand words numerically to identify unique words. It does not, however, allow a computer to understand meaning.

Imagine this scenario:

Tom loves to eat chocolate.

Tom likes to eat chocolate.

These sentences represented as a numerical array (list) would look like this:

1, 2, 3, 4, 5

1, 6, 3, 4, 5

As we can see, as humans both sentences are nearly identical. The only difference is the degree to which Tom appreciates eating chocolate. If we examine the numbers, however, these two sentences seem quite close, but their semantical meaning is impossible to know for certain. How similar is 2 to 6? The number 6 could represent “hates” as much as it could represent “likes”. This is where word vectors come in.

````python
import numpy as np
your_word = "kevin"

ms = nlp.vocab.vectors.most_similar(
    np.asarray([nlp.vocab.vectors[nlp.vocab.strings[your_word]]]), n=10)
words = [nlp.vocab.strings[w] for w in ms[0][0]]
distances = ms[2]
print(words)
````

The code uses spaCy's word vectors (from the `en_core_web_md` model) to find words semantically similar to a given word.

*   `your_word = "dog"`: This line defines the target word for which you want to find similar words.

*   `ms = nlp.vocab.vectors.most_similar(...)`: This is the core function call.
    *   `nlp.vocab.vectors.most_similar()`: This spaCy function takes a vector (or a list of vectors) and returns the most similar vectors in the vocabulary.
    *   `np.asarray([nlp.vocab.vectors[nlp.vocab.strings[your_word]]])`: This part retrieves the numerical vector representation of `your_word` ("dog"). `nlp.vocab.strings[your_word]` converts the string word into its numerical hash ID, which is then used to look up its vector in `nlp.vocab.vectors`. `np.asarray` ensures it's in the correct format for `most_similar`.
    *   `n=10`: This argument specifies that you want to retrieve the 10 most similar words.

*   `words = [nlp.vocab.strings[w] for w in ms[0][0]]`: The `most_similar` function returns a tuple containing three items. `ms[0]` is an array of word IDs and `ms[1]` is an array of similarity scores. `ms[0][0]` gives the IDs of the most similar words. This line then converts those numerical word IDs back into human-readable string words using `nlp.vocab.strings`.

*   `distances = ms[2]`: This extracts the similarity scores (or distances) corresponding to the `words` found. Higher scores typically indicate greater similarity.

*   `print(words)`: Finally, this line prints the list of words found to be most similar to "dog".

The `similarity()` method in spaCy is used to measure the semantic relatedness between two `Doc` objects, two `Span` objects, or two `Token` objects. It leverages **word vectors (word embeddings)** to do this.

In summary:

*   **Relies on Word Vectors**: For `similarity()` to work effectively, the loaded spaCy model must include word vectors (e.g., `en_core_web_md` or `en_core_web_lg`). Without them, the similarity score will often be 0 or an arbitrary value.
*   **Numerical Representation**: Each word (token) has a multi-dimensional numerical vector representing its meaning. The `similarity()` method compares these vectors.
*   **Cosine Similarity**: Internally, spaCy typically uses **cosine similarity** to calculate how 'alike' two vectors are. Cosine similarity measures the cosine of the angle between two vectors; a value close to 1 means they are very similar (small angle), and a value close to 0 means they are dissimilar (large angle).
*   **Aggregation for Documents/Spans**: When comparing documents or spans, spaCy typically aggregates the vectors of all the tokens within them (e.g., by averaging them) to create a single vector representation for the entire text unit. It then calculates the cosine similarity between these aggregated vectors.
*   **Semantic Relatedness**: The resulting similarity score (ranging from 0 to 1) indicates how semantically similar two texts or words are. For example, "dog" and "cat" would have a higher similarity score than "dog" and "building" because their meanings are more related.

# Adding Pipelines

In this instance, you would want to make a blank English model and simply add the Sentencizer to it. The reason is because each pipe in a pipeline will be activated (unless specified) and that means that each pipe from Dependency Parser to named entity recognition will be performed on your data. This is a serious waste of computational resources and time. The small model may take hours to achieve this task. By creating a blank model and simply adding a Sentencizer to it, you can reduce this time to merely minutes.

### Explanation of code in Cell `1ZvTvyOxCain`

*   **`import requests`**: This line imports the `requests` library, which is a popular Python library for making HTTP requests. It allows you to send HTTP/1.1 requests easily.

*   **`from bs4 import BeautifulSoup`**: This line imports the `BeautifulSoup` class from the `bs4` library (Beautiful Soup). Beautiful Soup is a Python library for pulling data out of HTML and XML files. It works with your favorite parser to provide idiomatic ways of navigating, searching, and modifying the parse tree.

*   **`s = requests.get("https://ocw.mit.edu/ans7870/6/6.006/s08/lecturenotes/files/t8.shakespeare.txt")`**: Here, `requests.get()` is used to send an HTTP GET request to the specified URL. The URL points to a plain text file containing Shakespeare's works. The response from this request (which includes the content of the file) is stored in the variable `s`.

*   **`soup = BeautifulSoup(s.content).text.replace("-\n", "").replace("\n", " ")`**: This is a multi-step operation:
    *   `BeautifulSoup(s.content)`: Creates a `BeautifulSoup` object from the content of the HTTP response (`s.content`). Even though it's a plain text file, `BeautifulSoup` can still process it. In this case, it helps in extracting the raw text cleanly.
    *   `.text`: Extracts all the text content from the `BeautifulSoup` object as a single string.
    *   `.replace("-\n", "")`: This method replaces occurrences of a hyphen followed by a newline character (common when words are hyphenated at the end of a line in text files) with an empty string. This effectively joins hyphenated words.
    *   `.replace("\n", " ")`: This method then replaces all remaining newline characters (`\n`) with a single space character. This converts the multi-line text into a continuous string, which is often preferred for NLP processing.
    *   The resulting cleaned text string is assigned to the variable `soup`.

*   **`nlp.max_length = 5278439`**: The `nlp` object here refers to the spaCy Language object that was previously initialized (in this notebook, it was `nlp = spacy.blank("en")` and `nlp.add_pipe("sentencizer")`).
    *   `nlp.max_length` is a setting in spaCy that defines the maximum number of characters a `Doc` object can contain. By default, spaCy has a relatively small `max_length` to prevent memory issues with extremely large texts.
    *   This line increases that limit to `5278439` characters. This is crucial because the Shakespeare text downloaded is quite large, and without increasing this limit, spaCy would raise an error when trying to process it, indicating that the text exceeds the default `max_length`.

In summary, this cell's purpose is to fetch a large text document from the web, clean it for consistent formatting, and configure the spaCy processor to handle its considerable size, preparing it for further natural language processing.
