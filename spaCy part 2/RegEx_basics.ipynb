# RegEx
Regular Expressions, or RegEx for short, is a way of achieving complex string matching based on simple or complex patterns. It can be used to perform finding and retrieving patterns or replacing matching patterns in a string with some other pattern.

# Working with RegEx

````python
import re
'''
\\d numeric digit between 0-9
{1,2} that occurs once or twice
a single space
| refers to either or
'''
pattern = r"((\d){1,2} (January|February|March|April|May|June|July|August|September|October|November|December))"
text = "This is a date 2 February. Another date would be 14 August."
matches = re.findall(pattern,text)
print(matches)
````

Limitations when date is reversed

````python
text = "This is a date February 2. Another date would be 14 August."
matches = re.findall(pattern, text)
print (matches)
````

````python
pattern = r"(((\d){1,2}( (January|February|March|April|May|June|July|August|September|October|November|December)))|(((January|February|March|April|May|June|July|August|September|October|November|December) )(\d){1,2}))"

text = "This is a date February 2. Another date would be 14 August."
matches = re.findall(pattern, text)
print (matches)
````

````python
iter_matches = re.finditer(pattern, text)
print (iter_matches)
for hit in iter_matches:
    print (hit)
````

````python
# Using re.findall (returns a list of tuples with captured groups)
print("Output from re.findall:")
print(re.findall(pattern, text))

print("\n---")

# Using re.finditer (returns an iterator of match objects)
print("Output from re.finditer (iterating through match objects):")
iter_matches = re.finditer(pattern, text)
for match in iter_matches:
    print(f"  Full match: '{match.group(0)}'")
    print(f"  Start index: {match.start()}")
    print(f"  End index: {match.end()}")
    print(f"  All captured groups: {match.groups()}")
    # You can access specific captured groups too, e.g., match.group(3) for the month in our pattern
    # The exact group index depends on your pattern's structure.
    print("\n")
````

There are multiple sets of parentheses, each creating a capturing group. When `re.findall` or `match.groups()` from `re.finditer` return results, they provide the content captured by each of these groups in the order they appear in the pattern, from left to right.
