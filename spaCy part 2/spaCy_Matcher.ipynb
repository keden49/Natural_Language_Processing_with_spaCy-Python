### spaCy Matcher

The `Matcher` in spaCy is a powerful, rule-based token matching engine that allows you to find sequences of tokens based on their lexical attributes within a `Doc` object. It is highly efficient and flexible, making it ideal for tasks like information extraction, pattern recognition, and custom entity annotation.

**Key Characteristics and Features:**

*   **Rule-Based Patterns:** You define patterns as a list of dictionaries, where each dictionary represents a token and its attributes (e.g., `TEXT`, `LEMMA`, `POS`, `TAG`, `DEP`, `ENT_TYPE`, `IS_ALPHA`, `IS_DIGIT`, `LIKE_URL`, etc.).
*   **Lexical and Linguistic Attributes:** Patterns can leverage a wide range of token attributes to create precise matching rules.
*   **Operators:** Special operators like `*` (zero or more), `+` (one or more), `?` (zero or one), and `!` (negation) can be used to specify the quantity and optionality of tokens in a pattern.
*   **Callbacks:** You can attach callback functions to patterns. When a match is found, the associated callback function is executed, allowing for custom processing or logic.
*   **Efficiency:** The `Matcher` is optimized for speed, even with many patterns and large texts.
*   **Use Cases:** Commonly used for identifying specific phrases, extracting structured information (e.g., dates, names following a specific format), or marking custom entities that don't fall under spaCy's built-in entity types.

**Basic Example of a Pattern:**

```python
[{"LOWER": "hello"}, {"IS_PUNCT": True, "OP": "?"}, {"LOWER": "world"}]
```
This pattern would match "hello world", "hello, world", "hello! world", etc.

### What does 'lexical' mean in NLP?

In natural language processing (NLP), 'lexical' refers to characteristics or properties related to individual words or tokens, as opposed to their grammatical structure or meaning within a larger context.

When we talk about 'lexical attributes' in spaCy's Matcher, we're referring to things like:

*   **The actual text of the word** (e.g., "run", "running")
*   **Its lowercase form** (e.g., "run" for both "Run" and "run")
*   **Whether it's alphanumeric, a digit, or punctuation**
*   **Its lemma** (the base form of the word, e.g., "run" for "runs", "running", "ran")

These are properties inherent to the word itself, rather than how it functions in a sentence.

Let's break down the code step by step to understand how the spaCy Matcher works and what each line accomplishes.

### **1. Setup: Importing spaCy and the Matcher**

```python
import spacy
from spacy.matcher import Matcher
```
*   **`import spacy`**: This line imports the main spaCy library, which provides access to all its functionalities, including language models and the `nlp` object.
*   **`from spacy.matcher import Matcher`**: This specifically imports the `Matcher` class from spaCy's `matcher` module. This is the tool we'll use to define and find patterns in text.

### **2. Loading the Language Model**

```python
nlp = spacy.load("en_core_web_sm")
```
*   **`nlp = spacy.load("en_core_web_sm")`**: This loads a small English language model (`en_core_web_sm`). This model is essential because it processes raw text into `Doc` objects, which contain linguistic annotations like tokens, part-of-speech tags, and entities. The `Matcher` operates on these `Doc` objects.

### **3. Initializing the Matcher**

```python
matcher = Matcher(nlp.vocab)
```
*   **`matcher = Matcher(nlp.vocab)`**: This line initializes an instance of the `Matcher`. We pass `nlp.vocab` to it because, as we discussed, the `Matcher` needs access to the shared vocabulary of the `nlp` object. This ensures that the patterns we define (which use lexical attributes like `LIKE_EMAIL`) are correctly mapped to the internal integer IDs that spaCy uses for efficient processing and consistency with the `Doc` objects generated by `nlp`.

### **4. Defining the Pattern**

```python
pattern = [{"LIKE_EMAIL": True}]
```
*   **`pattern = [{"LIKE_EMAIL": True}]`**: This is where we define the rule for what we want to find. A pattern is a list of dictionaries, where each dictionary describes a single token. Here, we have a single token dictionary: `{"LIKE_EMAIL": True}`.
    *   `"LIKE_EMAIL"`: This is a token attribute provided by spaCy. It evaluates to `True` if a token *looks like* an email address (e.g., contains an '@' symbol and a '.' followed by letters).
    *   `True`: This specifies that we are looking for a token where the `LIKE_EMAIL` attribute is `True`.
    *   So, this pattern will match any single token that spaCy identifies as an email address.

### **5. Adding the Pattern to the Matcher**

```python
matcher.add("EMAIL_ADDRESS", [pattern])
```
*   **`matcher.add("EMAIL_ADDRESS", [pattern])`**: This line adds our defined `pattern` to the `matcher` instance.
    *   `"EMAIL_ADDRESS"`: This is a unique string ID we give to this specific pattern. When the `Matcher` finds a match, it will return this ID, allowing us to know *which* pattern was matched.
    *   `[pattern]`: The second argument is a list of patterns. Even if you have only one pattern, it needs to be enclosed in a list.

### **6. Processing the Text**

```python
doc = nlp("This is an email address: kevinmotuka75@gmail.com")
```
*   **`doc = nlp("This is an email address: kevinmotuka75@gmail.com")`**: Here, we take our raw input string and pass it through the `nlp` object. The `nlp` object tokenizes the text, performs linguistic analysis (like POS tagging, dependency parsing, etc.), and returns a `Doc` object. The `Matcher` works on this `Doc` object.

### **7. Applying the Matcher and Getting Matches**

```python
matches = matcher(doc)
```
*   **`matches = matcher(doc)`**: This is the core step where the `Matcher` is applied. We call the `matcher` instance as a function, passing the `doc` object. The `Matcher` then iterates through the tokens in the `doc`, checking if any sequence of tokens matches the patterns it has stored (in our case, the `EMAIL_ADDRESS` pattern).

### **8. Printing the Results**

```python
print(matches)
print(nlp.vocab[matches[0][0]].text)
```
*   **`print(matches)`**: This prints the raw output from the `matcher`. The `matches` variable will be a list of tuples. Each tuple represents a single match and contains three elements:
    *   The **match ID** (a hash value of the pattern's string ID, e.g., `16571425990740197027` for "EMAIL_ADDRESS").
    *   The **start token index** of the match in the `doc`.
    *   The **end token index** of the match in the `doc`.
    *   In your output `[(16571425990740197027, 6, 7)]`, it means the pattern with ID `16571425990740197027` was found starting at token index 6 and ending at token index 7 (exclusive, so it's just token 6).

*   **`print(nlp.vocab[matches[0][0]].text)`**: This line retrieves the human-readable string ID of the matched pattern:
    *   `matches[0]`: Accesses the first (and in this case, only) match tuple `(16571425990740197027, 6, 7)`.
    *   `matches[0][0]`: Accesses the first element of that tuple, which is the **match ID** (the hash `16571425990740197027`).
    *   `nlp.vocab[...]`: We use the `nlp.vocab` to look up the string associated with this hash ID.
    *   `.text`: Retrieves the actual string value, which is "EMAIL_ADDRESS".

Let's break down the flow of the spaCy Matcher, specifically focusing on how patterns are processed and the role of `greedy='LONGEST'`.

Imagine the Matcher as a highly efficient detective looking for specific types of 'chains' (sequences of tokens) in your text. Here's a step-by-step process:

1.  **You define your 'blueprint' (the Pattern):**
    *   For our example, the pattern is `[{"POS":"PROPN", "OP":"+"}]`.
    *   This blueprint tells the detective:
        *   "I'm looking for a token that has the Part-of-Speech (POS) tag of **PROPN** (which stands for Proper Noun)."
        *   The `"OP":"+"` part means "I want **one or more** of these PROPN tokens appearing consecutively."

2.  **The text is already 'analyzed' (the `doc` object):**
    *   Before the Matcher even starts its work, spaCy's `nlp` object has already processed your raw text (`wiki_mlk.txt`). It has broken the text into individual tokens and, crucially, assigned a POS tag (like PROPN, VERB, NOUN, PUNCT) to each one. So, it knows 'Martin' is a PROPN, 'Luther' is a PROPN, 'King' is a PROPN, 'Jr.' is a PROPN, but '.' is PUNCT (punctuation), '(' is PUNCT, etc.

3.  **The detective starts scanning (calling `matcher(doc)`):**
    *   The `matcher` object (our detective) then systematically goes through your `doc` object, token by token, from the very beginning.

4.  **How the `PROPN+` pattern attempts to find a match:**
    *   **Finding a start:** The Matcher looks at the first token, `doc[0]` ('Martin'). It checks its POS tag. Is it a PROPN? Yes. Great! A potential sequence has started here.
    *   **Extending the sequence:** It then moves to `doc[1]` ('Luther'). Is it a PROPN? Yes. It continues: `doc[2]` ('King')? Yes. `doc[3]` ('Jr.')? Yes.
    *   **Terminating the sequence:** It then checks `doc[4]` ('('). Is it a PROPN? No, it's punctuation (PUNCT). At this point, the current sequence of consecutive PROPNs has ended.

5.  **This is where `greedy='LONGEST'` plays its role:**
    *   Once a sequence matching `PROPN+` has been identified (e.g., 'Martin Luther King Jr.'), the Matcher needs to decide how to record this match. Should it record 'Martin'? Or 'Martin Luther'? Or 'Martin Luther King'? Or 'Martin Luther King Jr.'?
    *   The `greedy='LONGEST'` setting instructs the Matcher: "When you find a valid sequence that matches the pattern starting at a particular token, make sure *that specific match* extends for as many tokens as possible before you declare it complete." # Collecting Multitokens ProperNouns

### spaCy Matcher Operators

| Operator | Description |
|----------|-------------|
| `!`      | Negate the pattern, by requiring it to match exactly 0 times. |
| `?`      | Make the pattern optional, by allowing it to match 0 or 1 times. |
| `+`      | Require the pattern to match 1 or more times. |
| `*`      | Allow the pattern to match 0 or more times. |
| `{n}`    | Require the pattern to match exactly n times. |
| `{n,m}`  | Require the pattern to match at least n but not more than m times. |
| `{n,}`   | Require the pattern to match at least n times. |
| `{,m}`   | Require the pattern to match at most m times. |

````python
matcher = Matcher(nlp.vocab)
pattern = [{"POS":"PROPN","OP":"+"}]
matcher.add("PROPER_NOUN", [pattern],greedy = "LONGEST")#configuring it to a specific pattern and a return string ma
doc = nlp(text) #tokenizes and identifies different attributes of text
matches = matcher(doc)#matches pattern to what nlp has identified

for match in matches[:10]:
  print(match, doc[match[1]:match[2]])#index at the second position reflects the index of the word in the doc object
````

````python
sorted_matches = sorted(matches, key=lambda x: x[1])

print("Top 10 matches sorted by their starting index:")
for match_id, start, end in sorted_matches[:10]:
    span = doc[start:end]
    print(f"({match_id}, {start}, {end}) {span.text}")
````

As you can see from the sorted output, `Martin Luther King Jr.` (starting at index 0) is indeed the first multi-word proper noun sequence found in the document. The previous output was simply a print of the unsorted list of matches.

````python
print(f"Token at index 88: '{doc[88].text}' (POS: {doc[88].pos_})")
print(f"Token at index 89: '{doc[89].text}' (POS: {doc[89].pos_})")
print(f"Token at index 90: '{doc[90].text}' (POS: {doc[90].pos_})")
````

# Finding Quotes and Speakers

````python
matcher = Matcher(nlp.vocab)
pattern = [{'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}]
matcher.add("QUOTES",[pattern],greedy='LONGEST')
doc = nlp(text)
matches = matcher(doc)
matches.sort(key = lambda x: x[1])
print (len(matches))
for match in matches[:10]:
    print (match, doc[match[1]:match[2]])
````
