This notebook seeks to demonstrate a stock analysis using spaCy on financial text. It covers the following steps:

1.  **Loading Data**: Imports stock symbols, company names, stock market indexes, and stock exchange information from TSV files.
2.  **Entity Ruler Creation**: Uses spaCy's `EntityRuler` to create custom patterns for identifying stocks, companies, indexes, and stock exchanges within text.
3.  **Text Analysis**: Applies the trained `EntityRuler` to a sample financial news article to extract and visualize the identified entities.

````python
#import main libraries
import spacy
import pandas as pd
````

````python
#load data into pandas dataframe

from google.colab import files as f
f.upload()
df = pd.read_csv("stocks.tsv", sep='\t')
````

````python
# Column descriptions for the DataFrame `df`:
# Symbol: A unique ticker symbol representing the company on the stock market.
# CompanyName: The full legal name of the company.
# Industry: The sector or industry in which the company primarily operates.
# MarketCap: The total market value of the company's outstanding shares.

df.head()
````

````python
symbols = df.Symbol.tolist()
companies = df.CompanyName.tolist()
print(symbols[:10])
````

````python
# importing indices
#f.upload()
df2 = pd.read_csv("indexes.tsv",sep = '\t')
df2.head()
````

````python
indexes = df2.IndexName.tolist()
index_symbols = df2.IndexSymbol.tolist()
````

````python

# importing stock exchanges
# BloombergExchangeCode: A unique code assigned by Bloomberg to identify a specific stock exchange.
# BloombergCompositeCode: A broader code that groups related exchanges or market segments within Bloomberg's system.
# Country: The country where the stock exchange is located.
# Description: A detailed description of the stock exchange.
# ISOMIC: The ISO 10383 Market Identifier Code, a unique identification code for exchanges, trading platforms, regulated markets, and trade reporting facilities.
# Google Prefix: A prefix used by Google Finance to identify stocks traded on this exchange.
# EODcode: An End-of-Day (EOD) code, likely used by data providers for referencing the exchange.
# NumStocks: The number of stocks listed on this particular exchange.
#f.upload()
df3 = pd.read_csv("stock_exchanges.tsv",sep ='\t')
df3.head()
````

````python
#generating a list of exchanges

exchanges = df3.ISOMIC.tolist() + df3["Google Prefix"].tolist() + df3.Description.tolist()
print(exchanges[:10])
````

````python
#creating entity ruler

nlp = spacy.blank("en")
ruler = nlp.add_pipe("entity_ruler")
patterns= []
letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

for symbol in symbols:
  patterns.append({"label": "STOCK", "pattern":symbol})
  for l in letters:
   patterns.append({"label": "STOCK", "pattern":symbol + f".{l}"})



stops = ["two"]
for company in companies:
  if company not in stops:
    patterns.append({"label": "COMPANY", "pattern":company})

for index in indexes:
  patterns.append({"label":"INDEX","pattern":index})
  words  = index.split()
  patterns.append({"label":"INDEX","pattern":" ".join(words[:2])})

for ids in index_symbols:
  patterns.append({"label":"INDEX","pattern":ids})

for e in exchanges:
  patterns.append({"label":"STOCK_EXCHANGES","pattern":e})
ruler.add_patterns(patterns)
````

````python
doc = nlp(text)
for ent in doc.ents:
  print(ent.text,ent.label_)
````
