# Introduction to spaCy entity ruler

The EntityRuler is a spaCy factory that allows one to create a set of patterns with corresponding labels. A factory in spaCy is a set of classes and functions preloaded in spaCy that perform set tasks. In the case of the EntityRuler, the factory at hand allows the user to create an EntityRuler, give it a set of instructions, and then use this instructions to find and label entities.

```python
import spacy
```

#Loading Model

```python
nlp = spacy.load("en_core_web_sm")
text = "West Chestertenfieldville was referenced in Mr. Deeds."
```

```python
doc  = nlp(text)
```

```python
for ent in doc.ents:
  print(ent.text,ent.label_)
```

```python
ruler = nlp.add_pipe("entity_ruler")
```

```python
nlp.analyze_pipes()
```


# Creating a pattern
Identifies pattern West Chestertenfieldville and assigns it a label of GPE
Earlier the model wrongly identified the entity as LOC since its never seen the data and also that the specific name is a fictional place in a movie

```python
patterns = [{"label":"GPE","pattern":"West Chestertenfieldville"},{"label":"FILM","pattern":"Mr. Deeds"}]
```

# Adding Patterns

```python
ruler.add_patterns(patterns)
```

# Testing it out

```python
doc2 = nlp(text)
for ent in doc2.ents:
  print(ent.text,ent.label_)
```

# Conclusion
Nothing Changed as NER pipeline comes before the newly defined entity ruler pipeline

# Rearranging order of pipelines

```python
nlp2 = spacy.load("en_core_web_sm")
```

```python
#placing the ruler pipe before the predefined ner
ruler = nlp2.add_pipe("entity_ruler",before="ner")
```

```python
ruler.add_patterns(patterns)
```

```python
doc3 = nlp2(text)
```

```python
for ent in doc3.ents:
  print(ent.text,ent.label_)
```

```python
nlp2.analyze_pipes()
```
