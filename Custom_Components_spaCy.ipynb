# Removing instances of GPE(GEOPOLITICAL ENTITY)

```python
from spacy.language import Language
#changes the doc object in the pipeline
@Language.component("remove_gpe")

def remove_gpe(doc):
  original_ents = list(doc.ents)#includes all entities even GPE

  for ent in doc.ents:
    if ent.label_ == "GPE":
      original_ents.remove(ent)

    doc.ents = original_ents
    return doc #doc that lacks GPE entities
```

```python
nlp.add_pipe("remove_gpe")
```

```python
nlp.analyze_pipes()
```
